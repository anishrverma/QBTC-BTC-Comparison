## Purpose
This notebook will be used to compare the performance of BTC-USD to the QBTC.U Fund.

## Things to troubleshoot
Using the pdr DataReader to extract Yahoo Finance Data for QBTC and BTC yields different number of dates for QBTC and BTC. Need to look at CSV to find out why.

```python
import pandas as pd
from datetime import date
import pandas_datareader.data as pdr
import matplotlib.pyplot as plt
```

```python
# Get and Save CSV Data
start = date(2020, 4, 20)
end = date.today()

BTC = pdr.DataReader('BTC-USD','yahoo',start,end)
QBTC = pdr.DataReader('QBTC-U.TO','yahoo',start,end)

BTC.to_csv('BTC-USD.csv')
QBTC.to_csv('QBTC-U.csv')
```

```python
# Plot Comparison of Close for USD
bDateList = BTC.index.tolist()
qDateList = QBTC.index.tolist()
bAdjClose = BTC['Adj Close'].tolist()
qAdjClose = QBTC['Adj Close'].tolist()

fig, ax1 = plt.subplots()
plt.xticks(rotation=45)

ax1.set_xlabel('Date')
ax1.set_ylabel('USD')
bAdjCloPlot = ax1.plot(bDateList, bAdjClose,'red', label = 'BTC-USD')

ax2 = ax1.twinx() 
ax2.set_ylabel('USD')  
qAdjCloPlot = ax2.plot(qDateList, qAdjClose, 'blue', label = 'QBTC-U.TO')

lns = bAdjCloPlot+qAdjCloPlot
labs = [l.get_label() for l in lns]
plt.legend(lns, labs, loc='center left', bbox_to_anchor=(1.1, 0.5))
plt.title('Adjusted Close Price (USD)')

plt.savefig('plots/Adjusted Close Comparison.pdf',bbox_inches='tight')
```

```python
plt.close()
```

```python
# Plot day-to-day percent change

bDateList = BTC.index.tolist()
qDateList = QBTC.index.tolist()
bAdjClose = BTC['Adj Close'].tolist()
qAdjClose = QBTC['Adj Close'].tolist()

qPerChange = [1]

for i in range(1,len(qAdjClose)-1):
    qPerChange.append(100*(qAdjClose[i]-qAdjClose[i-1])/qAdjClose[i-1])
#endfor

bPerChange = [1]

for i in range(1,len(bAdjClose)-1):
    bPerChange.append(100*(bAdjClose[i]-bAdjClose[i-1])/bAdjClose[i-1])
#endfor

plt.plot(bDateList[0:-1], bPerChange,'red', label = 'BTC-USD')
plt.plot(qDateList[0:-1], qPerChange,'blue', label = 'QBTC-U.TO')
plt.xticks(rotation=45)
plt.ylabel('Day-To-Day Percent Change (%)')
plt.xlabel('Date')
plt.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
plt.title('Day-to-Day Percent Change')

plt.savefig('plots/Day-to-Day Percent Change Comparison.pdf',bbox_inches='tight')
```
